## Merging multiple sets into single data 

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Jane', 'Alice', 'Bob', 'Charlie'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

df2 = pd.DataFrame({
    'employee_id': [2, 3, 4, 6, 7],
    'salary': [50000, 60000, 55000, 65000, 70000],
    'bonus': [5000, 6000, 5500, 6500, 7000]
})

df3 = pd.DataFrame({
    'department': ['HR', 'IT', 'Finance', 'Marketing'],
    'location': ['New York', 'San Francisco', 'Chicago', 'Boston']
})
```

```python
# inner merge


df_inner =  pd.merge( df1 , df2 , on='employee_id' ,  how='inner')

df_inner
```

```python
# left merge

df_left = pd.merge( df1 , df2 , on="employee_id" , how='left')

df_left
```

```python
# outer merge


df_outer =  pd.merge( df1 , df2 , on= "employee_id" ,  how="outer")

df_outer
```

```python
df_multiple = pd.merge( df_inner , df3 , on="department" ,  how="left")

df_multiple
```

```python
# mergring if they both have multiple names

df2_renamed = df2.rename(columns = { 'employee_id' : 'emp_id'})


df_merged_with_renamed = pd.merge( df2_renamed , df1 , left_on = 'emp_id' , right_on = "employee_id" ,  how="inner" )

df_merged_with_renamed
```
