## Concertination across axis

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                    'B': ['B0', 'B1', 'B2']},
                    index=['K0', 'K1', 'K2'])

df2 = pd.DataFrame({'C': ['C0', 'C1', 'C2'],
                    'D': ['D0', 'D1', 'D2']},
                    index=['K0', 'K2', 'K3'])

df3 = pd.DataFrame({'A': ['A3', 'A4', 'A5'],
                    'B': ['B3', 'B4', 'B5']},
                    index=['K3', 'K4', 'K5'])
```

```python
# vertically concatination - along rows

df_vert_concat =  pd.concat([ df1 ,  df3] )

df_vert_concat
```

```python
# horizontal concatinaton - along columns

df_hor_concat =  pd.concat([df1 , df2] , axis =  1)

df_hor_concat
```

```python
# concat with differenct indexes - filling missing values

df_outer = pd.concat([ df1 , df2 , df3 ] , axis = 1 ,  join='outer')

df_outer
```

```python
# concat with different indexes , only keep the intersections

df_inner =  pd.concat( [df1 , df2 , df3] , axis =1 ,  join='inner' )

df_inner
```

```python
# concat with index reset

df_reset =  pd.concat( [ df1 , df2 ] ,  ignore_index = True)

df_reset
```
