```python
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tickers = ['TSM' , 'NVDA']


df = yf.download(tickers, start="2024-01-01", end="2024-08-01", interval="1d")['Adj Close']
```

```python
df.head()
```

```python
# Plotting data in Histogram to see the distribution of the data

plt.hist(df['NVDA'], bins=20, alpha=0.5 )
plt.xlabel('Price')
plt.ylabel('Number of days observed')
plt.title('Histogram of NVDA prices in 2024')
plt.show()
```

```python
nvda_returns = df['NVDA'].pct_change().dropna()

plt.hist(nvda_returns, bins=20, alpha=0.5 )
plt.xlabel('Return')
plt.ylabel('Number of days observed')
plt.title('Histogram of NVDA returns in 2024')
plt.show()
```

```python
# Plotting returns distribution as a cummulative distribution

# it is useful to see the distribution of the returns in a cumulative way, as it gives a better idea of the distribution of the returns.

# it is best used in time series data

plt.hist(nvda_returns, bins=20, alpha=0.5, cumulative=True )
plt.xlabel('Return')
plt.ylabel('Number of days observed')
plt.title('Cumulative Histogram of NVDA returns in 2024')
plt.show()
```

```python
# scatter plot
plt.scatter(df['TSM'], df['NVDA'])
plt.xlabel('TSM')
plt.ylabel('NVDA')
plt.title('Scatter plot of TSM and NVDA prices')
plt.show()
```

```python
fig , axis = plt.subplots()

axis = sns.lineplot(data=df['NVDA'] , label='NVDA')
axis =  sns.lineplot(data=df['TSM'] , label = "TSM")

axis.set_xlabel('Date')
axis.set_ylabel('Price')
axis.set_title('TSM and NVDA prices in 2024')

plt.show()
```
