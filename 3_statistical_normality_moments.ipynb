## Statistical Analysis: Moments, Distributions, and Normality Testing

```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
"""
A distribution is statistically normal if it is symmetric and bell-shaped.

meaning that the distribution in both sides of the mean are mirror images of each other.

"""

# sample symetric and bell-shaped data

X = np.linspace(-5, 5, 1000)
Y = stats.norm.pdf(X, loc=0, scale=1)

plt.plot(X, Y)
plt.title('Normal Distribution')
plt.show()
```

```python
"""
Skewness in Distribution

Skewness measures the asymmetry of a distribution. 

Symmetric Distribution (Skewness = 0):

The left and right sides of the distribution are mirror images.
Mean = Median = Mode


Positively Skewed Distribution (Right-skewed, Skew > 0):

Has a longer tail on the right side.
Most values cluster on the left, with extreme values to the right.
Mean > Median > Mode


Negatively Skewed Distribution (Left-skewed, Skew < 0):

Has a longer tail on the left side.
Most values cluster on the right, with extreme values to the left.
Mean < Median < Mode

"""


xs2 = np.linspace(stats.lognorm.ppf(0.01, .7, loc=-.1), stats.lognorm.ppf(0.99, .7, loc=-.1), 150)

lognormal = stats.lognorm.pdf(xs2, .7)
plt.plot(xs2, lognormal, label='Skew > 0')


plt.plot(xs2, lognormal[::-1], label='Skew < 0')
plt.legend();
```

### Kurtosis

Kurtosis measures the "tailedness" of a probability distribution. It tells us about the shape of a distribution's peaks and tails compared to a normal distribution.

Types:

 - Mesokurtic (Normal): Kurtosis = 3
 - Leptokurtic: Kurtosis > 3, sharper peak, fatter tails
 - Platykurtic: Kurtosis < 3, flatter peak, thinner tails

Excess Kurtosis = Kurtosis - 3 (Normal distribution baseline)

```python
xs = np.linspace(-6,6, 300)
normal = stats.norm.pdf(xs)

plt.plot(xs,stats.laplace.pdf(xs), label='Leptokurtic')
print( 'Excess kurtosis of leptokurtic distribution:', (stats.laplace.stats(moments='k')))
plt.plot(xs, normal, label='Mesokurtic (normal)')
print( 'Excess kurtosis of mesokurtic distribution:', (stats.norm.stats(moments='k')))
plt.plot(xs,stats.cosine.pdf(xs), label='Platykurtic')
print( 'Excess kurtosis of platykurtic distribution:', (stats.cosine.stats(moments='k')))
plt.legend();
```
