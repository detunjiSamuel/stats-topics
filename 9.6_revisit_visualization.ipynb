```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df =  pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
```

```python
df.head()
```

```python
fig ,  axis = plt.subplots()
axis = sns.scatterplot( x = "total_bill" ,  y = "tip" ,  data=df ,  hue = "day"  , alpha = 0.8)
axis.set_title("Tip and total bill by day")
axis.set_xlabel("total bill")
axis.set_ylabel("tip")
fig.show()
```

```python
fig , axis = plt.subplots()

axis = sns.boxplot(x='day', y='total_bill', data=df)

axis.set_title( "Distribution of Total bill by Day")

plt.show()
```

```python
sns.pairplot(df , hue = "time")

plt.show()
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head(2)
```

```python
# histogram showing pertel lenght per species

fig , axis =  plt.subplots()


axis = sns.histplot(data=iris, x='sepal_length', hue='species',  kde=True, multiple='stack')

axis.set_title("Historgram showing the pertel lengh distribution per species")


fig.show()
```

```python
fig , axes = plt.subplots(2 , 2 , figsize =( 12 , 9))

fig.suptitle('Distributions of Iris Features', fontsize=16)

for ax, column in zip(axes.flatten(), iris.columns[:-1]):
    sns.histplot(data=iris, x=column, hue='species', kde=True, multiple='stack', ax=ax)
    ax.set_title(f'Distribution of {column.replace("_", " ").title()}')

plt.tight_layout()
plt.show()
```

```python
np.random.seed(0)
dates = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
data = pd.DataFrame({
    'date': dates,
    'value': np.cumsum(np.random.randn(len(dates))) + 100,
    'category': np.random.choice(['A', 'B', 'C'], len(dates))
})
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='date', y='value', data=data)
plt.title('Time Series Trend')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='date', y='value', hue='category', data=data)
plt.title('Time Series Trend by Category')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.legend(title='Category')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='date', y='value', data=data, ci=95)
plt.title('Time Series Trend with 95% Confidence Interval')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.show()
```

```python
tips = sns.load_dataset('tips')
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Average Total Bill by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Average Total Bill ($)')
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', hue='time', data=tips)
plt.title('Average Total Bill by Day and Time')
plt.xlabel('Day of the Week')
plt.ylabel('Average Total Bill ($)')
plt.legend(title='Time of Day')
plt.show()
```

```python
tips_grouped = tips.groupby(['day', 'time']).size().unstack()
tips_grouped_percentage = tips_grouped.div(tips_grouped.sum(axis=1), axis=0)

plt.figure(figsize=(10, 6))
tips_grouped_percentage.plot(kind='bar', stacked=True)
plt.title('Proportion of Lunch vs Dinner by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Proportion')
plt.legend(title='Time of Day')
plt.show()
```

```python
flights = sns.load_dataset("flights")

flights_pivot = flights.pivot("month", "year", "passengers")

plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, cmap="YlOrRd", annot=True, fmt="d")
plt.title("Number of Passengers per Month and Year")
plt.show()
```

```python
iris = sns.load_dataset("iris")
correlation = iris.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap="coolwarm", vmin=-1, vmax=1, center=0)
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.histplot(data=iris, x="sepal_length", y="sepal_width", cmap="YlGnBu")
plt.title("2D Histogram of Sepal Length vs Sepal Width")
plt.show()
```
